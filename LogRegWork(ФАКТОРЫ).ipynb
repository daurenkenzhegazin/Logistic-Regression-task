```python
#import libraries
import numpy as np
import pandas as pd


#import datasets
vars_ = pd.read_csv('vars-sample (2).csv',sep=';' , engine='python')
target = pd.read_csv('target-sample.csv', sep=';' , engine='python')
vars_.head()
```

```python
target.head()
```

## Clean Dataset

```python
vars_.info()
```

```python
#Check null values
vars_.isnull().sum()
```

```python
#Check null values
target.isnull().sum()
```

```python
#join dataset
final = vars_.merge(target,left_on='id', right_on='id', suffixes=(False, False))
final
```

```python
def ChangeFloat(column):
    list = []
    for index in column:
        index = index.replace(',','.')
        list.append(index)
    data = np.array(list)
    return data
```

```python
final['Factor_3'] = ChangeFloat(final['Factor_3']).tolist()
final['Factor_3'] = final['Factor_3'].astype(float)

final['Factor_4'] = ChangeFloat(final['Factor_4']).tolist()
final['Factor_4'] = final['Factor_4'].astype(float)

final['Factor_9'] = ChangeFloat(final['Factor_9']).tolist()
final['Factor_9'] = final['Factor_9'].astype(float)

final.info()
```

```python

final = final.replace('[^\d.]','',regex = True) 
```

```python
final.info()
```

```python
final['Factor_3'] = pd.to_numeric(final['Factor_3'],errors = 'coerce')
final['Factor_4'] = pd.to_numeric(final['Factor_4'],errors = 'coerce')
final['Factor_9'] = pd.to_numeric(final['Factor_9'],errors = 'coerce')
```

```python
final["Factor_3"] = final.Factor_3.astype(float)
final["Factor_4"] = final.Factor_4.astype(float)
final["Factor_9"] = final.Factor_9.astype(float)
```

```python
final.info()
```

```python
final.isnull()
```

```python
final = final.drop(['Factor_3','Factor_4','Factor_9'],axis =1)
```

```python
#create the X and Y
X = final.iloc[:,0:12]

y = final[['status']]
y = y.to_numpy()
```
